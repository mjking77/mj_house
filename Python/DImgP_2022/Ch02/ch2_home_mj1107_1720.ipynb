```python
import numpy as np
import cv2

global img

def onMouse( event, x, y, flags, param ):
	x, y = y, x
	if img.ndim != 3:
		print( "(x,y) = (%d, %d)" % (x, y), end = "  " )
		print( "Gray-Level = %3d" % img[x, y] )
	else:
		print( "(x,y) = (%d, %d)" % (x, y), end = "  " )
		print( "(R, G, B) = (%3d, %3d, %3d)" % 
			   ( img[x, y, 2], img[x, y, 1], img[x, y, 0] ) )		

filename = input( "Please enter filename: " )
img = cv2.imread( filename, -1 )
cv2.namedWindow( filename )
cv2.setMouseCallback( filename, onMouse )
cv2.imshow( filename, img )
cv2.waitKey( 0 )
```

```python
#ex_ch2-8
import cv2
import numpy as np

A = 200
T = 100
M = 512

img8 = np.zeros([512,512], dtype='uint8') #gray lv.
#img8 = np.zeros([100,512,3], dtype='uint8')   
for i in range(512):
    for j in range(512):
        img8[i][j]= A * np.sin(2*np.pi*i*T/M)     #gray lv
          
        
cv2.imshow("home1105",img8)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#ex_ch2-9
import cv2
import numpy as np

img9 = np.zeros([512,512], dtype='uint8') #gray lv.
#img8 = np.zeros([100,512,3], dtype='uint8')   

flag_inv = 0
for i in range(512):
    flag_inv = 0 if int((i//64)%2)==0 else 1  
    for j in range(512):        
        img9[i][j] = ((((j//64)%2) + flag_inv ) % 2)*255     
          
        
cv2.imshow("home1105",img9)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite("ch2_9.jpg",img9)
```

```python
#ex_ch2-10
import numpy as np
import cv2

filename = input( "Please enter filename: " )
img = cv2.imread( filename, -1 )
Rng_x, Rng_y = img.shape[:2]
cnt_1 = 5
roi_sz_x = 100
roi_sz_y = 100
roi_name = ""
img_ary = []

for i in range(cnt_1):
    ROI_x, ROI_y = eval( input( "Enter (x< %d, y<%d) for ROI: " %((Rng_x-roi_sz_x-1),(Rng_y-roi_sz_y-1)) ))   #use Str as express to Calc return
    #ROI_nr, ROI_nc = eval( input( "Enter (rows, columns) for ROI: " ) )
    ROI_nr, ROI_nc = roi_sz_x, roi_sz_y
    ROI = img[ ROI_x : ROI_x + ROI_nr, ROI_y : ROI_y + ROI_nc ]
    roi_name = "ROI_mj_" + str(i) + ".jpg"
    cv2.imwrite(roi_name, ROI)
    img_ary.append(cv2.imread(roi_name,-1))

img_all = np.hstack(img_ary)     #combine all pic show in together
cv2.imshow("home1107",img_all)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
